```python
mess = np.eye(24, k=1) + np.eye(24, k=-1) + np.eye(24, k=7) + np.eye(24, k=-7) + np.eye(24, k=8) + np.eye(24, k=-8)
mess += (np.eye(24, k=24 -7) + np.eye(24, k=-24 + 7) + np.eye(24, k=24 -8) + np.eye(24, k=-24 + 8))
mess[0,-1] = 1
mess[-1,0] = 1
plt.imshow(mess, cmap = 'gray')
plt.axis('off')
plt.show()

np.sum(mess, axis = 0)
```

```python
np.random.seed(1)
mat_maker = np.random.random(24)
mat_maker = mat_maker / np.sum(mat_maker)
mat_maker = mess * mat_maker.reshape(24,1)

#make mat_maker stochastic
mat_maker = mat_maker / np.sum(mat_maker, axis = 0)


eps = .1
#mat_maker = (mat_maker == 0) * eps/18 + mat_maker *(1 - eps)

print(np.sum(mat_maker, axis = 0))
plt.imshow(mat_maker, cmap = 'viridis')
```

```python
import matplotlib.pyplot as plt

# Sample data for demonstration (use your actual 'mean' and 'std' dictionaries)
methods = ['Method A', 'Method B', 'Method C']
mean = {'Method A': [0.8, 1.2], 'Method B': [0.9, 1.3], 'Method C': [0.85, 1.25]}
std = {'Method A': [0.1, 0.15], 'Method B': [0.2, 0.25], 'Method C': [0.05, 0.1]}
num_beta = 2  # Assuming you have two beta coefficients, beta_0 and beta_1

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(18, 5))  # Two subplots in one row

# Generate positions for each method on the x-axis
positions = range(len(methods))

# Loop over each beta
for i in range(num_beta):
    means = [mean[method][i] for method in methods]  # Extract means for beta_i
    stds = [2 * std[method][i] for method in methods]  # Extract std for beta_i and multiply by 2 for 2 std deviations
    
    # Plot each subplot with error bars
    ax[i].errorbar(positions, means, yerr=stds, fmt='o', capsize=5, label=f'Beta {i}')
    ax[i].set_xticks(positions)  # Set x-ticks to the positions
    ax[i].set_xticklabels(methods)  # Label x-ticks with method names
    ax[i].set_title(f'Estimates for Beta {i}')
    ax[i].set_xlabel('Method')
    ax[i].set_ylabel('Estimate')

plt.suptitle('Standard Deviation and Mean of Estimates')
plt.show()
```
